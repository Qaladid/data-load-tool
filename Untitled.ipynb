```python
def square_root_generator(limit):
    n = 1
    while n <= limit:
        yield n ** 0.5
        n += 1

# Example usage:
limit = 5
generator = square_root_generator(limit)

sqrt_values = []
for sqrt_value in generator:
    print(sqrt_value)
    sqrt_values.append(sqrt_value)

print(sum(sqrt_values))
```

```python
import dlt
import duckdb

def people_1():
    for i in range(1, 6):
        yield {"ID": i, "Name": f"Person_{i}", "Age": 25 + i, "City": "City_A"}

def people_2():
    for i in range(3, 9):
        yield {"ID": i, "Name": f"Person_{i}", "Age": 30 + i, "City": "City_B", "Occupation": f"Job_{i}"}

# Initialize DLT pipeline
pipeline = dlt.pipeline(destination='duckdb', dataset_name='person_data')

# Load data from the first generator into the table
first_one = pipeline.run(people_1(), table_name='people', write_disposition='replace')

# Establish a connection to DuckDB
conn = duckdb.connect(f"{pipeline.pipeline_name}.duckdb")
conn.execute(f"SET search_path = '{pipeline.dataset_name}'")

# sum of ages from the first generator
sum_of_age = conn.execute("SELECT SUM(Age) FROM people").fetchone()[0]
print("Sum of ages from first generator:", sum_of_age)


second_one = pipeline.run(people_2(), table_name='people', write_disposition='append')

# sum of ages of all people from both generators
age_total = conn.execute("SELECT SUM(Age) FROM people").fetchone()[0]
print("Sum of ages of all people after appending data from both generators:", age_total)
```

```python
import dlt

# Define generator functions
def people_1():
    for i in range(1, 6):
        yield {"ID": i, "Name": f"Person_{i}", "Age": 25 + i, "City": "City_A"}

def people_2():
    for i in range(3, 9):
        yield {"ID": i, "Name": f"Person_{i}", "Age": 30 + i, "City": "City_B", "Occupation": f"Job_{i}"}


pipeline = dlt.pipeline(destination='duckdb', dataset_name='merged_people')


first_load = pipeline.run(people_1(), table_name='people', write_disposition='replace', primary_key='ID')


second_load = pipeline.run(people_2(), table_name='people', write_disposition='merge')


conn = duckdb.connect(f"{pipeline.pipeline_name}.duckdb")
conn.execute(f"SET search_path = '{pipeline.dataset_name}'")


result = conn.execute("SELECT SUM(Age) FROM people")
total_age = result.fetchone()[0]
print("Sum of ages of all the people loaded:", total_age)
```
